```python
import pandas as pd
import matplotlib.pyplot as plt

import plotly.express as px
import plotly.io as pio
pio.renderers.default = "plotly_mimetype+notebook"
```

```python
df = pd.read_csv('D:\\TIL6022\\data\\Mobiliteitstrend__per_rit_en_motief_18102022_113630.csv',delimiter = ';')


df['Perioden'] = df['Perioden'].str.replace('*','')
#df = df.rename({'Afgelegde afstand per rit (gemiddeld) (km),,,,':'Afgelegde afstand per rit (gemiddeld) (km)'},axis = 'columns')
df['Ritten per persoon per dag (gemiddeld) (aantal)'] = df['Ritten per persoon per dag (gemiddeld) (aantal)'].str.replace(',','.').astype(float)
print(df.head())
df['Afgelegde afstand per rit (gemiddeld) (km)'] = df['Afgelegde afstand per rit (gemiddeld) (km)'].str.replace(',','.').astype(float)
#df.head()

#print(df.columns)

#print(df['Leeftijd'].value_counts())
#print(df['Reismotieven'].value_counts())
#print(df['Vervoerwijzen'].value_counts())
#print(df['Marges'].value_counts())
#print(df['Geslacht'].value_counts())

```

```python
df_1 = df[df['Leeftijd']=='Totaal']
df_1 = df_1[df_1['Reismotieven']=='Totaal']
df_1 = df_1[df_1['Vervoerwijzen']!='Totaal']

fig = plt.figure(figsize=(12, 9))
gs = fig.add_gridspec(nrows=2)
colors = ['blue','orange','green','red','purple','brown','black']


x = df_1['Perioden'].drop_duplicates()
y = df_1['Vervoerwijzen'].drop_duplicates()
labels = ['Car(driver)','Car(passenger)','Train','Bus/Tram/Metro','Bike','Walk','Other Modes']
ax = fig.add_subplot(gs[0])
j=0
for i in y:
    ax.plot(x,df_1[df_1['Vervoerwijzen']==i]['Ritten per persoon per dag (gemiddeld) (aantal)'],color = colors[j],marker = '.')
    j+=1
ax.set_xlabel('Year')
ax.set_ylabel('Trips per person per day (on Average)')
ax.set_xticks(ax.get_xticks()[::3])
ax.set_title('Trend of Trips per Day of Different Modes')
ax.legend(labels,loc='upper right',bbox_to_anchor=(1.23, 1))

ax = fig.add_subplot(gs[1])
j=0
for i in y:
    ax.plot(x,df_1[df_1['Vervoerwijzen']==i]['Afgelegde afstand per rit (gemiddeld) (km)'],color = colors[j],marker = '.')
    j+=1
ax.set_xlabel('Year')
ax.set_ylabel('Distance Travelled per Trip (on Average) (km)')
ax.set_xticks(ax.get_xticks()[::3])
ax.set_title('Trend of Distance Travelled per Trip of Different Modes')
ax.legend(labels,loc='upper right',bbox_to_anchor=(1.23, 1))

plt.show()



```
